```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import pickle

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
# Loading the dataset
cancer_data = load_breast_cancer()
```

```python
# Checking the attributes of the data
cancer_data.keys()
```

```python
# Checking the description of the data
print(cancer_data["DESCR"])
```

```python
# Printing the counts of target values
print(np.unique(cancer_data.target, return_counts=True))

# Displaying a message
print("\n")
print("Therefore, target 0 is Malignant and target 1 is Benign")
```

```python
# Creating the dataframe
cancer_df = pd.DataFrame(data=cancer_data["data"], columns=cancer_data["feature_names"])

# Displaying the dataframe
cancer_df.head()
```

### Exploratory Data Analysis

```python
# Checking for missing values
cancer_df.isna().sum()
```

```python
# Checking for correlation between features
cancer_df.corr()
```

### Data Preprocessing

#### I am first standardizing the data and then reducing the dimensions of the data using Principal Component Analysis. Then, I am finding which features were important enough to be considered while implementing PCA

```python
# Creating a standardiation object
scaler = StandardScaler()

# Standardizing the dataframe
scaled_data = scaler.fit_transform(cancer_df)

# Creating a dimensionality reduction object to reduce to 2 dimensions
pca = PCA(n_components=2)

# Reducing the dimensions of the data
principal_components = pca.fit_transform(scaled_data)

# Returning the reduced data
pd.DataFrame(principal_components).head()
```

```python
# Creating a dataframe to see the feature importance considered for PCA
pca_df = pd.DataFrame(pca.components_)

# Retrieving the locations of the most important features considered for PCA
loc_1 = pca_df.sort_values(by=0, ascending=False, axis=1).columns[0]
loc_2 = pca_df.sort_values(by=1, ascending=False, axis=1).columns[0]

# Retrieving the features present at the respective locations
feature_1 = cancer_df.columns[loc_1]
feature_2 = cancer_df.columns[loc_2]

# Displaying the feature names
print("The most important features used for PCA are '{}' and '{}'".format(feature_1, feature_2))
```

```python
# Plotting the reduced data
plt.scatter(principal_components[:,0], principal_components[:,1], c=cancer_data["target"])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Plotting The Principal Components")
```

```python
# Converting the reduced data into a dataframe
new_df = pd.DataFrame(principal_components, columns=[["PC1", "PC2"]])

# Adding the target values to new dataframe
new_df["Target"] = cancer_data["target"]

# Displaying the new dataframe
new_df.head()
```

### Model Training

```python
# Splitting the data into features and target
X = new_df.iloc[:,:-1].values
y = new_df.iloc[:,-1].values
```

```python
# Creating a function to perform Logistic Regression efficiently
def EfficientLogisticRegression(X, y):

    # Initializing max score to 0 and best model to Nonw
    max_score = 0
    best_model = None

    # Creating a for loop from 0 to 99
    for i in range(0, 100):

        # Splitting the data into train and test
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)

        # Creating a logistic regression model
        model = LogisticRegression()

        # Fitting the model
        model.fit(X_train, y_train)

        # Predicting the test targets
        y_pred = model.predict(X_test)

        # Calculating the accuracy of the model
        score = r2_score(y_test, y_pred)

        # Checking whether score is higher than max score
        if(score > max_score):

            # Replacing max score with score
            max_score = score

            # Replacing the best model with the current model
            best_model = model

    # Returning the max score and the model
    return max_score, model

# Rertieving the score and the model
logistic_score, logistic_model = EfficientLogisticRegression(X, y)

# Displaying the accuracy
print(logistic_score)
```

```python
# Creating a function to perform Support Vector Classification efficiently
def EfficientSVC(X, y):

    # Initializing max score to 0 and best model to Nonw
    max_score = 0
    best_model = None

    # Creating a for loop from 0 to 99
    for i in range(0, 100):

        # Splitting the data into train and test
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)

        # Creating a logistic regression model
        model = SVC()

        # Fitting the model
        model.fit(X_train, y_train)

        # Predicting the test targets
        y_pred = model.predict(X_test)

        # Calculating the accuracy of the model
        score = r2_score(y_test, y_pred)

        # Checking whether score is higher than max score
        if(score > max_score):

            # Replacing max score with score
            max_score = score

            # Replacing the best model with the current model
            best_model = model

    # Returning the max score and the model
    return max_score, model

# Retrieving the score and the model
svc_score, svc_model = EfficientSVC(X, y)

# Displaying the accuracy
print(svc_score)
```

```python
# Creating a function to perform Support Vector Classification efficiently
def EfficientNB(X, y):

    # Initializing max score to 0 and best model to Nonw
    max_score = 0
    best_model = None

    # Creating a for loop from 0 to 99
    for i in range(0, 100):

        # Splitting the data into train and test
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)

        # Creating a logistic regression model
        model = GaussianNB()

        # Fitting the model
        model.fit(X_train, y_train)

        # Predicting the test targets
        y_pred = model.predict(X_test)

        # Calculating the accuracy of the model
        score = r2_score(y_test, y_pred)

        # Checking whether score is higher than max score
        if(score > max_score):

            # Replacing max score with score
            max_score = score

            # Replacing the best model with the current model
            best_model = model

    # Returning the max score and the model
    return max_score, model

# Rertieving the score and the model
nb_score, nb_model = EfficientNB(X, y)

# Displaying the accuracy
print(nb_score)
```
